```python
import pandas as pd
```

```python
train = pd.read_csv("excel_train.csv")
train.head()
```

```python
validation = pd.read_csv("excel_validation.csv")
validation.head()
```

```python
test = pd.read_csv("excel_test.csv")
test.head()
```

```python
X_train = train.drop(['PassengerId','Survived'], axis = 1)
```

```python
y_train = train['Survived']
```

```python
X_val = validation.drop(['PassengerId','Survived'], axis = 1)
```

```python
y_val = validation['Survived']
```

#### Ideally we use this process to divide into train/test(validation in our case) 
- X = train.drop(['PassengerId', 'Survived'], axis = 1)
- y = train['Survived']
- import numpy as np
- from sklearn.model_selection import train_test_split
- X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=21)

```python
from sklearn.linear_model import LogisticRegression
```

```python
logModel = LogisticRegression(solver = 'lbfgs', max_iter = 1000)
```

```python
logModel.fit(X_train, y_train)
```

```python
logModel.score(X_train, y_train)
```

```python
logModel.coef_
```

```python
logModel.intercept_
```

### Evaluation of Validation

```python
X_val.head(1)
```

```python
y_val_pred = logModel.predict(X_val)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_val,y_val_pred ))
```

```python
#Close to what we have on our trained dataset
logModel.score(X_train, y_train)
```

### Prediction of Test

```python
X_test = test.drop(['PassengerId'], axis = 1)
```

```python
y_test_pred = logModel.predict(X_test).astype(int)
```

```python
AllSub = pd.DataFrame({ 'PassengerId': test['PassengerId'],
                       'Survived' : y_test_pred
    
})

AllSub.to_csv("All_Var_Video_Logistic.csv", index = False)
```
